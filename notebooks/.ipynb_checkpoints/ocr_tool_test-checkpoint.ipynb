# OCR Tool Test\nTest the docling-serve OCR tool on cropped table/figure regions from DocVQA documents.

````python
import sys
sys.path.insert(0, "/home/baris/repos/docvqa/src")

from PIL import Image
from IPython.display import display, Markdown
from datasets import load_dataset

Image.MAX_IMAGE_PIXELS = 500_000_000

from docvqa.solvers.flat_batch_solver import _ocr_via_docling

# Load dataset
ds = load_dataset("VLR-CVC/DocVQA-2026", split="val")
doc_map = {s["doc_id"]: s for s in ds}
print(f"Loaded {len(doc_map)} documents")
print("Available:", sorted(doc_map.keys()))
````

````python
def ocr_crop(page_img, box, label=""):
    """Crop a region, run OCR, display image and result side by side."""
    import tempfile
    crop = page_img.crop(box)
    w, h = crop.size
    
    # Run OCR
    tmp = tempfile.NamedTemporaryFile(suffix=".png", delete=False)
    crop.save(tmp, format="PNG")
    tmp.close()
    ocr_text = _ocr_via_docling(tmp.name)
    
    # Display
    title = f"**{label}** — crop {box}, {w}x{h}px" if label else f"Crop {box}, {w}x{h}px"
    display(Markdown(f"---\n### {title}"))
    display(crop)
    display(Markdown(f"**Raw OCR output:**\n```\n{ocr_text}\n```"))
    display(Markdown(f"**Rendered:**\n\n{ocr_text}"))
    return ocr_text
````
